```python

embeddings = OpenAIEmbeddings(model="text-embedding-3-small", chunk_size=400)
```

```python
db = FAISS.from_documents(final_documents,embeddings)
db
```

```python
query = "what is langchain"
docs = db.similarity_search(query)
docs
```

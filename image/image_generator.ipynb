```python
import os
from langchain_community.utilities.dalle_image_generator import DallEAPIWrapper
from langchain_core.prompts import PromptTemplate
from langchain_openai import OpenAI
from langchain_classic.chains import LLMChain
```

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ['OPENAI_API_KEY']=os.getenv("OPENAI_API_KEY")
os.environ['LANGCHAIN_API_KEY']=os.getenv("LANGCHAIN_API_KEY")
os.environ['LANGCHAIN_TRACING_V2']='true'
os.environ['LANGCHAIN_PROJECT']="Image App 2"

os.environ['LANGSMITH_API_KEY']=os.getenv("LANGCHAIN_API_KEY")
os.environ['LANGSMITH_TRACING']='true'
os.environ['LANGSMITH_PROJECT']="Image Apps 2"
os.environ['LANGSMITH_ENDPOINT']="https://api.smith.langchain.com"


```

```python

# Ensure API key is set in environment variables
# os.environ["OPENAI_API_KEY"] = "your_api_key_here"

# 1. Define the LLM to process the initial request
llm = OpenAI(temperature=0.7)

# 2. Create a prompt template for the LLM to generate a detailed image description
prompt_template = PromptTemplate(
    input_variables=["image_desc"],
    template="Generate a detailed prompt to create an image based on the following description of 16:9 ratio: {image_desc}",
)

# 3. Create the LLM chain
chain = LLMChain(llm=llm, prompt=prompt_template)

# 4. Use the DallEAPIWrapper as the final step
# user_description = "a spooky Halloween night at a haunted museum"
user_description = "A charming, quaint Swiss village of Mürren perched on a mountain ledge, traditional wooden houses with dark roofs, colorful flower boxes, dramatic Eiger, Mönch, and Jungfrau mountains in the background, autumn colors, cozy, peaceful, 8k, photorealistic"
user_description ="generate a Breathtaking Landscapes (Mountains, Fjords, Rainforests) image"
detailed_prompt = chain.run(user_description)
print(f"Detailed prompt: {detailed_prompt}")
```

```python

from langchain_community.utilities.dalle_image_generator import DallEAPIWrapper

prompt = "photorealistic, candid portrait of a given description with natural skin texture and slight imperfections, standing in a [setting]. Soft natural lighting, shot on 85mm lens, f/1.8, shallow depth of field, creamy background bokeh, 8k resolution, documentary photography style, Description : "
dalle_tool = DallEAPIWrapper(model="dall-e-3",size="1792x1024") # Specify dall-e-3
image_url = dalle_tool.run(prompt+detailed_prompt)
print(image_url)
```
